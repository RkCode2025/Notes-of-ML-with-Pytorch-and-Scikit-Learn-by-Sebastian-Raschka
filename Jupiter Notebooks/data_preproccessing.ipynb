**DECISION TREES WITH GINI AND ENTROPY**

```python
import numpy as np
from io import StringIO
import pandas as pd
```

```python
def entropy(p):
    return - p * np.log2(p) - (1- p) * np.log2((1-p))
```

```python
x = np.arange(0.0, 1.0, 0.01)
y = entropy(x)
print(y)
```
